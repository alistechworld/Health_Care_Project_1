```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.histplot(df["Blood Type"], kde = True, color = "Red")
plt.title("Distribution of Numeric Column")
plt.show()

# Box plot for visualization outliers
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Blood Type"])
plt.title("Box plot of Numeric Columns")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
blood_type_counts = df["Blood Type"].value_counts()
plt.pie(blood_type_counts, labels=blood_type_counts.index, colors=['red', 'blue', 'green', 'orange', 'black', 'yellow', 'aqua'], 
        autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
plt.title("Distribution of Blood Types")
plt.show()
```

```python
# Step 1: Select only the numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Step 2: Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Step 3: Plot the heatmap of the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
